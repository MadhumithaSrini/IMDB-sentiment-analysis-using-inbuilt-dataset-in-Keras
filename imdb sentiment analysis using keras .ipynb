```python
#importing all the packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
from tensorflow.python.keras.datasets import imdb
(x_train, y_train), (x_test,y_test) = imdb.load_data(num_words = 10000)
```

```python
print(x_train[1])
```

```python
print(x_test[1])
```

```python
word_index = imdb.get_word_index()
print(word_index["hey"])   #here in the bag of words, word is the key and the token is the value
```

```python
# here the word is changed into value and the token is changed into key, so that it will be easy to decode the words using their keys(tokens)
to_get_the_word_from_indices = dict((value,key) for key,value in word_index.items())
```

```python
print(to_get_the_word_from_indices[1397])
```

```python
# to decode an entire review
def dec(any_review):
    actual_text = ""
    for i in any_review:
        actual_text+=to_get_the_word_from_indices[i]
        actual_text+=" "
    return actual_text
```

```python
dec(x_train[1])
```

```python
from keras.preprocessing import sequence

#in order to make all the reviews of equal length by padding 
x_train = sequence.pad_sequences(x_train, value=word_index["a"], padding = "post" ,maxlen=200)
x_test = sequence.pad_sequences(x_test, value=word_index["a"], padding = "post", maxlen=200)
```

```python
dec(x_train[1])
```

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Embedding, GlobalAveragePooling1D

model = Sequential([Embedding(10000, 16), 
                   GlobalAveragePooling1D(),
                   Dense(16,activation = "relu"),
                   Dense(1,activation = "sigmoid")])

model.compile(
    loss = "binary_crossentropy",
    optimizer =  "adam",
    metrics = ["accuracy"])

model.summary()
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=2, verbose=1)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print("Loss is : ",loss)
print("Accuracy is : ",accuracy*100)
```

```python
def prediction(arg):
    Positive = 0
    Negative = 0
    for i in range(len(arg)):
        sentiment = ""
        actual = model.predict(np.expand_dims(arg[i], axis = 0))
        if(actual>0.5):
            sentiment = "Positive"
            Positive+=1
        else:
            sentiment = "Negative"
            Negative+=1

        print(actual," ",sentiment )
```

```python
prediction(x_test)
```

```python
#let us consider the first ten reviews
if Negative>Positive:
    print("OOPS:( The overall review is negative")
else:
    print("Hurray! :) The overall review is positive")
```
